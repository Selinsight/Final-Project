```python
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
columns_to_drop = ['diet', 'ethnicity', 'last_online', 'location', 'religion', 'sign','pets']
columns_to_drop = [col for col in columns_to_drop if col in df.columns]
df.drop(columns=columns_to_drop, inplace=True)
```

```python
categories = df.orientation.unique()
print(categories)
```

```python
def merge_gender_orientation(row):
    if row['orientation'] in ['gay', 'bisexual']:
        return 'other'
    return row['sex']

# Creating a new column 
df['gender'] = df.apply(merge_gender_orientation, axis=1)

# Dropping the sex and orientation columns
df = df.drop(['sex','orientation'], axis=1)

print(df['gender'].value_counts())
```

```python
# Identifying non-integer columns
non_integer_columns = df.select_dtypes(exclude=['int64', 'float64']).columns

# Replacing NaN values in non-integer columns with "no comment"
df[non_integer_columns] = df[non_integer_columns].fillna("no comment")

# Verify that NaN values are replaced
print(df[non_integer_columns].isnull().sum())  
```

```python
def simplify_education(education):
    if "high school" in education:
        return "High School"
    elif "college/university" in education or "two-year college" in education:
        return "College"
    elif "masters" in education:
        return "Masters"
    elif "ph.d" in education:
        return "Ph.D."
    elif "law school" in education or "med school" in education:
        return "Professional"
    else:
        return "Other"  # Includes Space Camp and other rare categories

# Apply the function to the education column
df['education_simplified'] = df['education'].apply(simplify_education)

# Drop the original column if no longer needed
df = df.drop(['education'], axis=1)

# Check the new simplified column
print(df['education_simplified'].value_counts())
```

```python
def kids(offspring):
    if "doesn't have kids" in offspring:
        return "No Kids"
    elif "has kids" in offspring or "has a kid" in offspring:
        return "Has Kids"
    elif "wants kids" in offspring or "might want kids" in offspring:
        return "Wants Kids"
    elif "doesn't want kids" in offspring:
        return "Doesn't Want Kids"
    else:
        return "No Comment"

# Apply the function to the offspring column
df['kids'] = df['offspring'].apply(kids)

# Drop the original column if no longer needed
df = df.drop(['offspring'], axis=1)

# Check the new simplified column
print(df['kids'].value_counts())
```

```python
def smoking(smokes):
    if smokes in ['no', 'no comment']:
        return "Non-Smoker"
    elif smokes in ['sometimes', 'when drinking']:
        return "Occasional Smoker"
    elif smokes == 'yes':
        return "Regular Smoker"
    elif smokes == 'trying to quit':
        return "Quitting Smoker"
    else:
        return "Unknown"  # In case of unexpected values

# Apply the function to the smokes column
df['smoking'] = df['smokes'].apply(smoking)

# Drop the original column if no longer needed
df = df.drop(['smokes'], axis=1)

# Check the new simplified column
print(df['smoking'].value_counts())
```

```python
print(df['speaks'].value_counts())
```

```python
def simplify_languages(speaks):
    if pd.isna(speaks):
        return "Unknown"
    
    # Split the languages and fluency levels
    languages = speaks.lower().split(", ")
    
    # Identify fluently spoken languages
    fluent_languages = [lang for lang in languages if "fluently" in lang]
    
    # Categorize based on the number of fluent languages
    if len(fluent_languages) == 1 and "english" in fluent_languages[0]:
        return "English Fluent Only"
    elif len(fluent_languages) == 2:
        return "Bilingual Fluent"
    elif len(fluent_languages) >= 3:
        return "Multilingual Fluent"
    elif any("english" in lang and ("poorly" in lang or "okay" in lang) for lang in languages):
        return "Basic English Speaker"
    else:
        return "Unknown"

# Apply the function to the speaks column
df['language'] = df['speaks'].apply(simplify_languages)

# Drop the original column if no longer needed
df = df.drop(['speaks'], axis=1)

# Check the new language_category column
print(df['language'].value_counts())
```
